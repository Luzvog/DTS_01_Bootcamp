### 15) El Teorema CAP dice:<br>
1- Que un sistema SQL tiene la característica de ser Consistente.<br>
2- Que un sistema No-SQL puede ser Tolerante a Particiones y Consistente.<br>
3- Que garantizar al mismo tiempo Consistencia, Disponibilidad y Tolerancia a Particiones no es posible para ningún sistema de bases de datos.<br>

3- Que el sistema disponibilizar información aún habiendo tenido fallas de red.<br>
